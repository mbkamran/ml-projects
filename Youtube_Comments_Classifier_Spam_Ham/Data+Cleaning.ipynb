```python
import pandas as pd
import numpy as np
import glob
```

```python
csv_files = glob.glob("Youtube*.csv")
df_list = (pd.read_csv(file) for file in csv_files)
yt_df = pd.concat(df_list, ignore_index=True)

twt_df = pd.read_csv("train.csv")
```

```python
# a = twt_df[['Tweet', 'Type']]
ready_twt = (twt_df.rename(columns={'Tweet': 0, 'Type': 1}))[[0, 1]]
ready_twt.loc[ready_twt[1] == 'Quality', 1] = 0
ready_twt.loc[ready_twt[1] == 'Spam', 1] = 1

ready_twt
```

```python
ready_twt = ready_twt.sample(1956)
```

```python
ready_yt = (yt_df.rename(columns={'CONTENT': 0, 'CLASS': 1}))[[0, 1]]
ready_yt
```

```python
ready_yt = ready_yt.sample(1000)
```

```python
stopwords = pd.read_csv("stopwords.csv")
stopwords
```

```python
df_data = pd.concat([ready_twt, ready_yt])
df_data
```

```python
import re

df_data[0] = df_data[0].str.lower()
df_data[0] = df_data[0].str.replace(r'@[a-z0-9]+', ' TAGGEDUSER ', regex=True)
df_data[0] = df_data[0].str.replace(r'#[a-z0-9]+', ' HASHTAG ', regex=True)
df_data[0] = df_data[0].str.replace(r'(https?:\/\/)?[\w\-~]+(\.[\w\-~]+)+(\/[\w\-~@:%]*)*(#[\w\-]*)?(\?[^\s]*)?', ' HTTP ', regex=True)
df_data[0] = df_data[0].str.replace(r'(^\d+\s|\s\d+\s|\s\d+$)|(\d*\.\d+)|(\d+\.[0-9 ]+)', ' NUMBER ', regex=True)
df_data[0] = df_data[0].apply(lambda x: ' '.join(w for w in x.split() if not w in ["&amp", "&gt", "&lt"]))
df_data[0] = df_data[0].apply(lambda x: ' '.join(w for w in x.split() if not w in stopwords))
df_data[0] = df_data[0].apply(lambda x: re.sub('[^A-Za-z0-9]', ' ', x))
df_data[0] = df_data[0].apply(lambda x: ' '.join(w for w in x.split() if len(w)>1))
```
